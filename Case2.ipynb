```r
#load library
library(dplyr)
library(ggplot2)
library(tidyverse)
```

# NO. 1

```r
#Read data
data = read.table("auto-mpg.data")
cat("Jumlah row = ",nrow(data))
data
```

```r
#Clean data with NA value
data_dropna = na.omit(data)
cat("Jumlah row = ",nrow(data_dropna))
data_dropna
```

```r
#Check duplicate row
data_duplicate = data_dropna[duplicated(data_dropna),]
data_duplicate
```

```r
#Clean data duplicate row
data_drop_duplicate = data_dropna[!duplicated(data_dropna),]
data_drop_duplicate
```

```r
str(data_drop_duplicate)
```

# NO. 2

```r
df = data_drop_duplicate
# density = density(df$V5)
# density

#Draw plot for density of V5
density = ggplot(df, aes(x=V5)) + geom_density() + labs(title = "V5 Density")
density
```

# NO. 3

```r
ggplot(df, aes(sample = V10, colour = abbreviate(V10))) + stat_qq()
```

# NO. 4 & NO. 5

```r
#Draw linear regression of V5
ggplot(df, aes(x=V5, y=V5)) + geom_point() + stat_smooth(method = "lm", col = "red")
```
