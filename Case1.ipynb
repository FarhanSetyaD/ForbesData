# CASE 1

```r
#read data
df = read.csv('data_Forbes.csv')
# df = df[,c(3,6)]
df
```

# 1.1

```r
#create dataframe with country name is United States
us = df[df$country == "United States",]
# us

#create dataframe United States coloumn only profits
us_profit = select(us,profits)
# typeof(us_profit)

#calculate and print the median using median()
cat('median profit of us = ', median(unlist(us_profit), na.rm = 'True'))
```

```r
#create dataframe with country name is United Kingdom, France, and Germany
multi_country = filter(df, (country == "United Kingdom") | (country == "France") | (country == "Germany"))

#create dataframe multi_country coloumn only profits
multi_country_profit = select(multi_country,profits)

#calculate and print the median using median()
cat('median profit of multi country = ', median(unlist(multi_country_profit), na.rm = 'True'))
```

# 1.2

```r
#Find all Germany Companies with negative profit
filter(df, (country == "Germany") & (profits < 0))
```

# 1.3

```r
#Create dataframe with country is Bermuda
bermuda = filter(df, (country == "Bermuda"))
bermuda
```

```r
#Count all category frequent
find = table(select(bermuda, category))

#Find max category
max = which.max(find)
max
```

```r
#Count companies that have insurance category
nrow(filter(df, (category == rownames(find)[max])))
```

```r
#Sorting 20 highest profit
sort = head(arrange(df,desc(profits)), n = 20)
sort
```

```r
#Plotting the data
ggplot(data = sort, mapping = aes(x=marketvalue, y=profits, color=abbreviate(country))) + geom_point()
```
